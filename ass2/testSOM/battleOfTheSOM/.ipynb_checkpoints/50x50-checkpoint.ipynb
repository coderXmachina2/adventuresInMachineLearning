```python
#50 x 50 u matrix implement
import matplotlib.pylab as plt
import pandas as pd
import numpy as np
import sys
import sompy
from time import time
from sompy.visualization.bmuhits import BmuHitsView

np.set_printoptions(threshold=sys.maxsize) #see all elements, dont truncate
```

```python
dFrame = pd.read_csv('somDatasetLLM.csv')
quakeLLMRecArr  = dFrame.to_records(index=False)
quakeLLM = np.array(quakeLLMRecArr.tolist())
```

```python
quakeLLM
```

```python
mapsize = [50,50] #define map size
fiftyFiftySom = sompy.SOMFactory.build(quakeLLM, 
                                       mapsize, 
                                       mask=None, 
                                       mapshape='planar', 
                                       lattice='rect', 
                                       normalization='var', 
                                       initialization='pca', 
                                       neighborhood='gaussian', 
                                       training='batch', 
                                       name='sompy') 

# this will use the default parameters, but I can change the initialization and neighborhood methods

fiftyFiftySom.train(n_job=1, 
                      shared_memory = 'no', 
                      verbose='debug')  # verbose='debug' will print more,
                            # verbose=None wont print anything
```

```python
v = sompy.mapview.View2DPacked(6, 6, 'Visualization',text_size=12) 
v
```

```python
fiftyFiftySom.component_names = ['Longitude','Latitude','Magnitude']
v.show(fiftyFiftySom, what='codebook', which_dim='all', cmap=None, col_sz=6) #which_dim='all' default
```

```python
v.show(fiftyFiftySom, what='codebook', which_dim='all', cmap='rainbow', col_sz=4) #which_dim='all' default
```

```python
h = sompy.hitmap.HitMapView(10, 10, 'hitmap', text_size=10, show_text=True)
h.show(fiftyFiftySom)
```

```python
fiftyFiftySom.project_data(quakeLLM) # we can project each data point to a neuron
```

```python
u = sompy.umatrix.UMatrixView(12, 
                              12, 
                              'umatrix', 
                              show_axis=True, 
                              text_size=8, 
                              show_text=True) #prep the plot

#This is the Umat value
UMAT  = u.build_u_matrix(fiftyFiftySom, 
                         distance=1,  
                         row_normalized=False) #build the plot

#Here you have Umatrix plus its render
UMAT = u.show(fiftyFiftySom, 
              distance2=1, 
              row_normalized=False,
              show_data=True, 
              contooor=True, 
              blob=True)
```

```python
greyMat = UMAT[1]
plt.imshow(greyMat, cmap="gray")
plt.suptitle('Umatrix of 50x50 SOM', fontsize=24)
plt.show()
```

```python
#do this towards the end. See the map first then assign cluster Labels
vis2 = sompy.mapview.View2DPacked(2, 2, 'test',text_size=8)  
#first you can do clustering. Currently only K-means on top of the trained som
cl = fiftyFiftySom.cluster(n_clusters=5) #this is cluster labels
# print cl
getattr(fiftyFiftySom, 'cluster_labels')
```

```python
map_labels = fiftyFiftySom.cluster(n_clusters=5)
data_labels = np.array([map_labels[int(k)] for k in fiftyFiftySom._bmu[0]]) # mapping labels from grid to original data
```

```python
map_labels
```

```python
data_labels
```

```python
countClust0= 0
countClust1= 0
countClust2= 0
countClust3= 0
countClust4= 0

for t in range(0, len(data_labels)):
    if (data_labels[t] == 0):
        countClust0 += 1
    elif (data_labels[t] == 1):
        countClust1 += 1
    elif (data_labels[t] == 2):
        countClust2 += 1
    elif (data_labels[t] == 3):
        countClust3 += 1
    elif (data_labels[t] == 4):
        countClust4 += 1
        
clustCount = np.array([countClust0, countClust1, countClust2, countClust3, countClust4]) #sizes
```

```python
print("cluster 0: " + str(countClust0))
print("cluster 1: " + str(countClust1))
print("cluster 2: " + str(countClust2))
print("cluster 3: " + str(countClust3))
print("cluster 4: " + str(countClust4)) # this is where the magnitude is
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'cluster 0', 'cluster 1', 'cluster 2', 'cluster 3', 'cluster 4 * where the epicenter is at'
#sizes = [15, 30, 45, 10]
explode = (0, 0, 0, 0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(clustCount , explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.suptitle('Distribution of datapoints to Cluster', fontsize=24)

plt.show()
```

```python
clusteredOneSevOneSevLLM = np.column_stack((quakeLLM, data_labels))
np.savetxt("clusteredx50LLM.csv", clusteredOneSevOneSevLLM , fmt='%.20f' , delimiter=",")
```
