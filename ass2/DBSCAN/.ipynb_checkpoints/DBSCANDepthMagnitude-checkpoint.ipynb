```python
#dbscanMagnitude

import matplotlib.pylab as plt
import pandas as pd
import numpy as np
import sys
import sompy
from time import time
from sompy.visualization.bmuhits import BmuHitsView
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN

np.set_printoptions(threshold=sys.maxsize) #see all elements, dont truncate
```

```python
dFrame = pd.read_csv('somDatasetLLM.csv')
quakeRecArr  = dFrame.to_records(index=False)
quakeLLM  = np.array(quakeRecArr.tolist()) #either a dont delete it or adaptfurther down.
```

```python
quakeLLM
```

```python
from sklearn.neighbors import NearestNeighbors

ns = 4
nbrs = NearestNeighbors(n_neighbors=ns).fit(quakeLLM)
distances, indices = nbrs.kneighbors(quakeLLM)
distanceDec = sorted(distances[:,ns-1], reverse=True)
```

```python
figure = plt.figure()
plt.plot(list(range(1,3291)), distanceDec)
figure.set_size_inches(12,12)
plt.axvline(x=76, color = 'green', alpha=0.2, markersize=2)
plt.axhline(y=0.135, color = 'green', alpha=0.2, markersize=2)

plt.axvline(x=38, color = 'red', alpha=0.2, markersize=2)
plt.axhline(y=0.112, color = 'red', alpha=0.2, markersize=2)
```

```python
clustering = DBSCAN(eps=0.135, min_samples=2.2).fit(quakeLLM)
clustering.labels_
```

```python
print("cluster labels range from: " + str(min(clustering.labels_)) + " " + str(max(clustering.labels_)))
```
