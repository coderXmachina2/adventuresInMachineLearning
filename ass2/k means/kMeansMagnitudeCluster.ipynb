```python
import matplotlib.pylab as plt
import pandas as pd
import numpy as np
import sys
import sompy
from time import time
from sompy.visualization.bmuhits import BmuHitsView
from sklearn.cluster import KMeans

np.set_printoptions(threshold=sys.maxsize) #see all elements, dont truncate
```

```python
dFrame = pd.read_csv('somDatasetLLM.csv')#its in the same directory
#dFrame = pd.read_csv('clusteredx50LLD.csv') #17x17 resulted yielded 4 clusters so now 
quakeLLMRecArr  = dFrame.to_records(index=False)
quakeLLM = np.array(quakeLLMRecArr.tolist())
```

```python
quakeLLM #go 3d k means
```

```python
def plotKMeanScatter(kMeanTable , kMeanModel):
    from sklearn.cluster import KMeans  
    fig = plt.figure()
    ax = fig.add_subplot(111)

    
    ax.scatter(kMeanTable[:,0], kMeanTable[:,1],c=kMeanModel.labels_, cmap='jet', alpha =0.5,  edgecolor='k')
    ax.scatter(kMeanModel.cluster_centers_[:,0] ,kMeanModel.cluster_centers_[:,1], cmap='jet')  
    #fig.savefig('matplotlibs\Plot.png')
    
    plt.suptitle('kmeans plot', fontsize=24)
    
    fig.set_size_inches(10,6)
    
    plt.xlim(171, 174)
    plt.ylim(-44, -42.5)
    
    plt.grid()
    
    ax.set_xlabel('$longitude$')
    ax.set_ylabel('$latitude$')
    
    plt.grid(c='blue')
    plt.show()
```

```python
import time
start = time.time()

kmean = KMeans(n_clusters=4) #from previously we find idal clusters is 4
kmean.fit(quakeLLM )

end = time.time()
print(end - start)

```

```python

kmean = KMeans(n_clusters=4) #from previously we find idal clusters is 4
kmean.fit(quakeLLM )


plotKMeanScatter(quakeLLM, kmean)
```
